```python
import re
from apify_client import ApifyClient

# Initialize the ApifyClient with your API token
client = ApifyClient("apify_api_RnnezEXgDhvVWxvHJzdMoMqP2QGkz81LZWOf")

# Generate all possible pairs of hashtags
pairs = list(combinations(hashtag_counts.keys(), 2))

# Calculate the combined count for each pair
pairs_with_counts = [(pair, hashtag_counts[pair[0]] + hashtag_counts[pair[1]]) for pair in pairs]

# Sort the pairs by their combined count in descending order
sorted_pairs = sorted(pairs_with_counts, key=lambda x: x[1], reverse=True)

# Get the top 5 pairs with the highest count
top_5_pairs = sorted_pairs[:5]

# Function to extract Instagram handle from URL
def extract_instagram_handle(url):
    match = re.search(r'instagram\.com/([^/]+)', url)
    if match:
        return match.group(1)
    return None

# Set to keep track of unique handles
unique_handles = set()

# Iterate through the top 5 pairs and run the Apify Actor for each pair
for pair in top_5_pairs:
    hashtag1, hashtag2 = pair[0]
    query = f'instagram.com "{hashtag1}" "{hashtag2}"'

    # Prepare the Actor input
    run_input = {
        "queries": query,
        "resultsPerPage": 5,
        "maxPagesPerQuery": 1,
        "languageCode": "",
        "mobileResults": False,
        "includeUnfilteredResults": False,
        "saveHtml": False,
        "saveHtmlToKeyValueStore": False,
        "includeIcons": False,
    }

    # Run the Actor and wait for it to finish
    run = client.actor("nFJndFXA5zjCTuudP").call(run_input=run_input)

    # Fetch and print Actor results from the run's dataset (if there are any)
    for item in client.dataset(run["defaultDatasetId"]).iterate_items():
        for result in item.get('organicResults', []):
            handle = extract_instagram_handle(result['url'])
            if handle and handle not in unique_handles and len(handle)>1 and handle != "reel" and handle!= profile:
                unique_handles.add(handle)
                print(handle)
```
